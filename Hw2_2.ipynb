```python
plt=reload(plt)
diag2 = python_data.groupby('Посада')['Зарплата на місяць'].agg(['mean'])
diag2 = diag2.reset_index()
plt.xticks(rotation = 45, color = 'purple',fontsize = 'small')
plt.xlabel('Proffesions', fontsize = 15)
plt.ylabel('Mean salary', fontsize = 15)
plt.title("Розаподіл зарплати за професіями")
plt.bar(diag2['Посада'],diag2['mean'])
```

```python
plt=reload(plt)
diag3 = python_data.copy(deep=True)
diag3["Загальний досвід роботи"] = pd.to_numeric(diag3["Загальний досвід роботи"], errors="coerce")
diag3.dropna(subset=["Загальний досвід роботи"], inplace=True)
diag3 = diag3.groupby('Посада')['Загальний досвід роботи'].agg(['mean'])
diag3 = diag3.reset_index()
plt.plot(diag3['Посада'],diag3['mean'])
plt.xticks(rotation = 90)
plt.title("Salary of professions")
plt.xlabel("Professions")
plt.ylabel("Salary")
```

```python
group_data = python_data.groupby('Посада')['Зарплата на місяць'].agg(['count','min','max','sum'])
```

```python
group_data = group_data.reset_index()
group_data
```

 **13**

```python
def fill_avg_salary(row):
  return row['sum']/row['count']

group_data['avg'] = group_data.apply(fill_avg_salary,axis = 1)
```
