```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import patoolib
import os
```

```python
patoolib.extract_archive('homework.zip')
```

```python
path = os.listdir('/content/data')
full_df = pd.DataFrame()
temp = 0
for i in path:
    for j in os.listdir('/content/data/' + i):
        path_to_scv = '/content/data/' + i + '/' + j
        df = pd.read_csv(path_to_scv)
        df['min X'] = np.min(df.iloc[:,0])
        df['min Y'] = np.min(df.iloc[:,1])
        df['min Z'] = np.min(df.iloc[:,2])
        df['std X'] = np.std(df.iloc[:,0])
        df['std Y'] = np.std(df.iloc[:,1])
        df['std Z'] = np.std(df.iloc[:,2])
        df['mean X'] = np.mean(df.iloc[:,0])
        df['mean Y'] = np.mean(df.iloc[:,1])
        df['mean Z'] = np.mean(df.iloc[:,2])
        df['max X'] = np.max(df.iloc[:,0])
        df['max Y'] = np.max(df.iloc[:,1])
        df['max Z'] = np.max(df.iloc[:,2])
        df['median X'] = np.median(df.iloc[:,0])
        df['median Y'] = np.median(df.iloc[:,1])
        df['median Z'] = np.median(df.iloc[:,2])
        df['class'] = i
        df['class_index'] = temp
        full_df = pd.concat([full_df,df])
    temp += 1
```

```python
full_df
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X1 = full_df.iloc[:,:-2]
y1 = full_df.iloc[:,-1]
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.33, random_state=42)
clf1 = svm.SVC()
clf1.fit(X_train1, y_train1)
y_pred1 = clf1.predict(X_test1)
print(accuracy_score(y_test1,y_pred1))
print(classification_report(np.array(y_test1), y_pred1))
```

```python
X2 = full_df.iloc[:,0:3]
y2 = full_df.iloc[:,-1]
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.33, random_state=42,shuffle=True)
clf2 = svm.SVC()
clf2.fit(X_train2, y_train2)
y_pred2 = clf2.predict(X_test2)
print(accuracy_score(y_test2,y_pred2))
print(classification_report(np.array(y_test2), y_pred2))
```

**Як ми бачимо, чим менше фіч, за якими навчається модель, тим гірші результати метрик, а що найбільш цікаве - усіх з них. Тому додавання більшої кількості різноманітних фіч дає змогу моделі краще та швидше навчитися.**

```python
X3 = full_df.iloc[:,:-2]
y3 = full_df.iloc[:,-1]
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.33, random_state=42)
rndfrst1 = RandomForestClassifier(max_depth=2, random_state=0)
rndfrst1.fit(X_train3, y_train3)
y_pred3 = rndfrst1.predict(X_test3)
print(accuracy_score(y_test3,y_pred3))
print(classification_report(np.array(y_test3), y_pred3))
```

```python
X4 = full_df.iloc[:,0:3]
y4 = full_df.iloc[:,-1]
X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y4, test_size=0.33, random_state=42)
rndfrst2 = RandomForestClassifier(max_depth=2, random_state=0)
rndfrst2.fit(X_train4, y_train4)
y_pred4 = rndfrst2.predict(X_test4)
print(accuracy_score(y_test4,y_pred4))
print(classification_report(np.array(y_test4), y_pred4))
```

**Як і у минулому випадку, ми бачимо, що більша кількість фіч впливає на більш гарні результати усіх метрик.**

**Якщо порівнювати ці дві моделі, можна побачити, що SVC алгоритм має більш точні результати майже з усіх метрик, на однаковій кількості фіч, на яких навчалась модель. Значення avg взагалі набагато гірші. Це може бути пов'язано з обраною кількість "глибини", та інших гіперпараметрів, на яких було створено модель.**
